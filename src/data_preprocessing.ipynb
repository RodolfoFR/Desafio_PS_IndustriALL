**Object to Write Important Information to a Text File:**

**Getting the Path of CSV files from Raw Data Folder:**

```python
# path of raw_data folder
folder_csv_files = os.path.join(path_of_PS_IndustriALL, "PS_IndustriALL/raw_data")

# sorted list of complete path of the csv files
path_csv_files:list = [os.path.join(folder_csv_files, file) for file in sorted(os.listdir(folder_csv_files)) if file.endswith(".csv")]

# Putting the output csv file in the first position of the csv file list, for organization
path_output_csv_file = path_csv_files.pop()
path_csv_files.insert(0, path_output_csv_file)
```

**Load the Raw Data in Pandas Dataframe:**
  - raw_data (pandas.core.frame.DataFrame) : The Dataframe/Spreadsheet of the Raw Data

```python
raw_data = pd.read_csv(path_csv_files[0]) # first column is "timestamp", second is the Output (target_iALL_PS)
output_column = raw_data.columns[1]
removed_inputs = list() # list of removed inputs
for path in path_csv_files[1:]: # loop through each feature csv file

  feature_df = pd.read_csv(path) # read the csv file

  # check if the feature csv file has the same column of another feature csv file or if have empty column
  for raw_data_column in raw_data.columns[1:]:
    for feature_df_column in feature_df.columns[1:]:
      if ((feature_df[feature_df_column] == raw_data[raw_data_column]).all()) or (feature_df[feature_df_column].isna().all()):
        feature_df.drop(columns=feature_df_column, inplace=True)
        removed_inputs.append(feature_df_column)

  if len(feature_df.columns) > 1:
    raw_data = pd.merge(raw_data, feature_df, how="inner", on="timestamp")


# Annotation of information
note_taker.write_line("Amount the Raw Data", len(raw_data))
note_taker.write_line("Inputs", len(raw_data.columns)-2)
note_taker.write_line("Amount Removed Inputs", len(removed_inputs))
note_taker.write_line("Removed Inputs", removed_inputs)
note_taker.write_line("Output", 1)

display(raw_data) # display the raw_data DataFrame
```

**Removing the Possible Duplicate Data:**

```python
data = raw_data.drop_duplicates(subset="timestamp") # remove the possible duplicate data based on "timestamp" column
note_taker.write_line("Duplicate Data", len(raw_data)-len(data)) # annotation of amount duplicate data
```

**Removing The Data With Empty Inputs:**

```python
data.dropna(inplace=True)
note_taker.write_line("Amount the Complete Data", len(data))
display(data.tail())
```

**Check the Possible Data Outputs:**

```python
possible_outputs = data.iloc[:,1].unique()
note_taker.write_line("Possible Outputs", possible_outputs) # annotation of amount duplicate data
data_per_type_output = list() # list the DataFrames for each output
for output in possible_outputs:
  data_per_type_output.append(data[data.iloc[:,1] == output])

for i in range(len(data_per_type_output)):
  note_taker.write_line(f"Porcent of Data with Output {possible_outputs[i]}", f"{(len(data_per_type_output[i])/len(data))*100:.1f}%")
```


**Plot of Graphics:**
  - The Graphics are Input x Time
  - Each Color of the Graphics represent one Output

```python
for column in data.columns[2:]:
  plt.figure(figsize=(15,6))
  plt.plot(data_per_type_output[0][column], color='blue', label=possible_outputs[0])
  plt.plot(data_per_type_output[1][column], color='red', label=possible_outputs[1])
  plt.xlabel('time')
  plt.ylabel(column)
  plt.title(column)
  plt.grid(True)
  plt.legend()
  plt.show()
```

**Data analysis:**

  - Mean, standard deviation  (std)  and Coefficient of Variation (cv) of Each Column Data

```python
input_informations = dict() # dictionary to store information about inputs

for column in data.columns[2:]: # Each input column of Dataframe

  # mean of data form column
  averages = [data_per_type_output[0][column].mean(), data_per_type_output[1][column].mean()]

  # std of data form column
  std = [data_per_type_output[0][column].std(), data_per_type_output[1][column].std()]

  # cv of data form column
  cv = [(std[0]/averages[0])*100, (std[1]/averages[1])*100 ]

  # outpout 0 dictionary
  output_0 = dict()
  output_0["mean"] = averages[0]
  output_0["std"] = std[0]
  output_0["cv"] = cv[0]

  # outpout 1 dictionary
  output_1 = dict()
  output_1["mean"] = averages[1]
  output_1["std"] = std[1]
  output_1["cv"] = cv[1]

  # column dictionary
  input_informations[column] = dict()
  input_informations[column][possible_outputs[0]] = output_0
  input_informations[column][possible_outputs[1]] = output_1

  input_informations_string = f"[({averages[0]:.2f} / {averages[1]:.2f}), ({std[0]:.2f} / {std[1]:.2f}), ({cv[0]:.2f}% / {cv[1]:.2f}%)]"

  # annotation
  note_taker.write_line(f"{column} (Normal/Anormal) [mean, std, cv]", input_informations_string)
  print("="*110)
```

  - Checking the Smaller Coefficients of Variation
  - Inputs with Coefficient of Variation less than (mean - standard deviation) are recommended to be removed from the Dataset

```python
# calculate the mean, standard deviation and the threshold of columns Coefficients of Variation
cv = list()
for column in data.columns[2:]:
  cv.append(input_informations[column][possible_outputs[0]]["cv"])

cv_mean = np.mean(cv) # mean of coefficients of variation
cv_std = np.std(cv) # std of coefficents of variation

# Threshold coefficent of variation value accepted
cv_threshold = cv_mean - cv_std

# annotation
note_taker.write_line("Mean of Coefficients of Variation from inputs", f"{cv_mean}%")
note_taker.write_line("Standard deviation of Coefficients of Variation from inputs", f"{cv_std}%")
note_taker.write_line("Threshold of Coefficients of Variation from inputs", f"{cv_threshold}%")

# check which column has the coefficients of Variation lower than threshold
input_informations["low_cv"] = list()
for column in data.columns[2:]:
  if input_informations[column][possible_outputs[0]]["cv"] <= cv_threshold:
    input_informations["low_cv"].append(column)
    note_taker.write_line(f"Input with Coefficients of Variation lower than threshold", column)
```

  - Create the Hisograms:
  - Each Column will have be 2 histograms
  - 1 for each output (Normal or Anormal)

```python
histograms = dict() # dictionary with all histograms
for column in data.columns[2:]: # for in each input columns
  histograms[column] = dict()
  for i in range(len(possible_outputs)):
    histogram = Histogram(data_per_type_output[i][column], column)
    histograms[column][possible_outputs[i]] = histogram
```

  - Plot Histograms per Columns:

```python
for column, histograms_per_column in histograms.items():
    for output, histogram in histograms_per_column.items():
      color = "blue" if output == possible_outputs[0] else "red"
      histogram.plot(3, 3, color=color,  title=f"{column} ({output})", top=0.2)
```

  - Calculating the relative intersection area between the same column's histograms
  - Inputs with intersection area of histograms (Normal/Anormal) greater than 75% are recommended to be removed from the Dataset
  - Because inputs with this condition above has low variation in classes (Normal/Anormal)

```python
threshold = 0.75 # threshold of intersection area

input_informations["big_intersections"] = list() # list of inputs with intersection area greater than threshold

for column, histograms_per_column in histograms.items(): # Each input column

  histogram_class_0 = histograms_per_column[possible_outputs[0]] # histogram of class 0 (Normal)
  histogram_class_1 = histograms_per_column[possible_outputs[1]] # histogram of class 1 (Anormal)

  # relative intersection area between the histograms from column
  relative_intersection_area = histogram_class_0.intersection(histogram_class_1)

  # Store The information of relative intersection area from column
  input_informations[column]["relative_intersection_area"] = relative_intersection_area

  # annotation
  note_taker.write_line(f"{column} with relative intersection area of histograms", relative_intersection_area)
  print("="*100)

  # check if column's relative intersection area is greater than threshold, if True store in input_informations (dict)
  if relative_intersection_area >= threshold:
    input_informations["big_intersections"].append(column)
# annotation
note_taker.write_line("big intersections", input_informations["big_intersections"])
```

**Inputs to Remove:**
  - Criterion:
      + Low Coefficient of Variation
      + intersection area of histograms (Normal/Anormal) greater than 75%
  - The Inputs with this Criterion have little influence on the output


```python
# list of inputs(columns) that will be removed
removed_columns = list(set(input_informations["low_cv"]) & set(input_informations["big_intersections"]))

# Removing the inputs(columns) of data (pd.DataFrame)
data.drop(columns=removed_columns, inplace=True)

# annotation
note_taker.write_line("Inputs/Columns Removed based on criterion ", removed_inputs)
```

**Conversion the Output's Type:**
  - Conversion String Type to Float Type
  - (Normal/Anormal) to (0.0, 1.0)

```python
dict_to_replace_output = dict() # dictionary to change the output column
for new_output, old_output in enumerate(possible_outputs): # The new output will be index (0 or 1) of possible_outputs (list)

  dict_to_replace_output[old_output] = float(new_output) # filling the dictionary with new output type

data[output_column] = data[output_column].replace(dict_to_replace_output) # changing the output column
```

**Data Correlation:**

```python
data_correlation = data.corr()
display(data_correlation)
```

**Choosing the Best Minute for Predict:**

  - Criterion of range of minutes wiil validation to be evalualed:
    + 30 min before the possible anomaly
    + 480 min (8h) before the possible anomaly, because it is equivalent to a working day the worker
    + The minute has the best mean of absolute correlations will be choose

```python
min_predict_minute = 30 # 30 min
max_predict_minute = 8 * 60  # 8h
shifted_data = data.copy() # copy of data (pd.Dataframe)

best_mean = 0
best_minute = 0

list_minutes_mean = list() # store the means

for minute in range(min_predict_minute, max_predict_minute+1): # each minute of minutes range

  minute_info = dict() # store the mean of this minute
  minute_info["minute"] = minute
  # shifting the output values to the line with the prediction minute
  shifted_data[output_column] = data[output_column].shift(-minute)

  shifted_data.dropna(inplace=True) # remove the line with NaN values

  amount_shifted_data = len(shifted_data)

  # correlation between the prediction minute inputs and the output it want to predict
  correlation = shifted_data.corr()
  correlation_with_output = correlation.drop(output_column)[output_column]

  # Mean of absolute correlations from inputs with output
  correlation_mean = correlation_with_output.abs().mean()

  minute_info["mean"] = correlation_mean
  list_minutes_mean.append(minute_info) # store the info abbout this minute

  # store the best minute and mean
  if best_mean < correlation_mean:
    best_mean = correlation_mean
    best_minute = minute

  # print per minute
  print(f"minute: {minute}")
  print(f"mean: {correlation_mean}")
  print(f"amount shifted data: {amount_shifted_data}")
  print("="*30)

# annotation
note_taker.write_line("max mean", best_mean)
note_taker.write_line("minute", best_minute)
```

**Selected the minute to predict:**
  - The selected minute has mean 97% equal or greater than best minute

```python
tolerance_minute = dict() # store the tolerance minute (dict), which is the selected minute
for minute in list_minutes_mean: # Each minute

  if minute["minute"] == best_minute:
    continue

  if minute["mean"] > 0.97 * best_mean:
    tolerance_minute["minute"] = minute["minute"]
    tolerance_minute["mean"] = minute["mean"]

note_taker.write_line("Minute of tolerance", tolerance_minute)
```

**Plot of correlation of selected Minute:**

```python
shifted_data = data.copy()
shifted_data[output_column] = data[output_column].shift(-tolerance_minute["minute"]).dropna()
correlation_shifted_data = shifted_data.corr()
correlation_shifted_data = correlation_shifted_data.drop(output_column).loc[:, [output_column]]
display(correlation_shifted_data)
```





**Removing The Inputs (Columns) with Correlation with output low than that mean of selected minute**

```python
removed_columns = list() # columns to remove
mean = tolerance_minute["mean"] # threshold of correlation
for column, corr in correlation_shifted_data[output_column].items():
  if abs (corr) <  mean:
    removed_columns.append(column)

# Removing the inputs(columns) of data (pd.DataFrame)
data.drop(columns=removed_columns, inplace=True)
note_taker.write_line("Romoved Columns with low Correlation with Output", removed_columns)
note_taker.write_line("Number of inputs after preprocessing", data.columns-2)
print("Final Data Preporcessing:")
display(data)
```

**Based on the main model (which predicts the occurrence of anomalies in the industrial plant), another prediction model will be created to predict how long the anomaly will last**
  - This model is used after the main model prediction
  - This anomaly time prediction happens 63 minutes (previously determined) before the detection of a possible anomaly

**Data Preprocessing of new Model**

- Recovery of original data
- With basic cleaning
- Conversion String Type to Float Type on target_iALL_PS	column

```python
data = raw_data.drop_duplicates(subset="timestamp") # remove the possible duplicate data based on "timestamp" column

dict_to_replace_output = dict() # dictionary to change the output column
for new_output, old_output in enumerate(possible_outputs): # The new output will be index (0 or 1) of possible_outputs (list)

  dict_to_replace_output[old_output] = float(new_output) # filling the dictionary with new output type

data[output_column] = data[output_column].replace(dict_to_replace_output) # changing the output column

display(data)
```

**Creation of the new Output**
  - Anomaly Time

```python

anomaly_time = list()
previous_stage = 0
sequence_counter = 0

for id_line, row in data.iterrows(): # each line of data (dataframe)

  current_stage = row["target_iALL_PS"]

  if (current_stage == previous_stage) and (current_stage == 0): # normality
    sequence_counter = 0
    anomaly_time.append(0)
    previous_stage = current_stage


  elif (current_stage != previous_stage) and (current_stage == 1): # emergence of anomaly

    sequence_counter = 1
    anomaly_time.append(0)
    previous_stage = current_stage

  elif (current_stage == previous_stage) and (current_stage == 1): # continuation of the anomaly

    sequence_counter += 1
    anomaly_time.append(0)
    previous_stage = current_stage

  elif (current_stage != previous_stage) and (current_stage == 0): # end of the anomaly

    # fill in the anomaly sequence
    for _ in range(sequence_counter):
      anomaly_time[-sequence_counter] = sequence_counter
      sequence_counter -= 1

    sequence_counter = 0
    anomaly_time.append(0)
    previous_stage = current_stage

# Conversion of List to pd.Series
anomaly_time = pd.Series(anomaly_time)
# insertion of new column
data.insert(1, "anomaly_time", anomaly_time )

# Remove the data with Nan/None, this could break the sequence, but it is necessary
data.dropna(inplace=True)
display(data)
```

**Cheking the Correlation of this New Dataset**





```python
data_correlation = data.corr()
display(data_correlation)
```

```python
shifted_data = data.copy()
shifted_data["anomaly_time"] = data["anomaly_time"].shift(-tolerance_minute["minute"]).dropna()
correlation_shifted_data = shifted_data.corr()
correlation_shifted_data = correlation_shifted_data.drop("anomaly_time").loc[:, ["anomaly_time"]]
display(correlation_shifted_data)
```


**Calculte the mean of Correlations with anomaly time and Removing The Inputs (Columns) with Correlation with anomaly time low than that mean**

```python
# Mean of absolute correlations from inputs with output
correlations_mean = correlation_shifted_data.abs().mean()[0]

removed_columns = list() # columns to remove

for column, corr in correlation_shifted_data["anomaly_time"].items():
  if abs(corr) <  correlations_mean:
    removed_columns.append(column)

# Removing the inputs(columns) of data (pd.DataFrame)
data.drop(columns=removed_columns, inplace=True)
note_taker.write_line("Romoved Columns with low Correlation with anomaly time", removed_columns)

print("Final Data Preporcessing of new Model:")
display(data)
```
